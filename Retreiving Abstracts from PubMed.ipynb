# Retreiving Abstracts from PubMed

A simple example on how to extract articles from PubMed Central using E-Utilities provided by NCBI. Details on how to use E-Utilities can be found here:(http://www.ncbi.nlm.nih.gov/books/NBK25499/)

This example returns 2 abstracts of articles in text form with PubMed ID 17284678 and 9997.

```python
import urllib.request
url = "http://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed&id=17284678,9997&retmode=text&rettype=abstract"
response = urllib.request.urlopen(url)
data = response.read()      # a `bytes` object
text = data.decode('utf-8') # a `str`; this step can't be used if data is binary
```

```python
print(text)
```

You can refine your search by add search terms and filter, for example, I am only interested in English journal articles where the title or abstract contains the term miRNA or microRNA or microRNA, and has free access to the full text.

```python
import urllib.request
url = "http://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&term=asthma&WebEnv=<webenv string>&usehistory=y&retmode=xml&rettype=abstract"
response = urllib.request.urlopen(url)
data = response.read()      # a `bytes` object
text = data.decode('utf-8') # a `str`; this step can't be used if data is binary
```

```python
print(text)
```

The search criteria used in this the example found in https://github.com/ryubidragonfire/EntitiesRelationExtraction/ looks like this:

`((miRNA[Title/Abstract] OR microRNA[Title/Abstract] OR micro RNA[Title/Abstract]) AND "english"[Language]) AND "journal article"[Publication Type] AND "loattrfree full text"[sb]`
